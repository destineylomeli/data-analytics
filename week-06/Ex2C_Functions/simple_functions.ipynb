# Favorite Things

```python
def favorite_things(name, movie, music):
    return f"Name: {name}\nFavorite Movie: {movie}\nFavorite Musician/Band: {music}"
```

```python
print(favorite_things("Destiney", "Coraline", "Alice in Chains"))
```

# Favorite Place

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."
```

```python
print(favorite_place("the beach", "I enjoy hearing the waves along with watching the sunset it all feels so peaceful to me"))
```
