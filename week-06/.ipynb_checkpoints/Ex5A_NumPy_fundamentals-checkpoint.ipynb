# Exercise 5.A NumPy Fundamentals

This notebook practices creating and working with NumPy arrays,
including one dimensional arrays, two dimensional arrays,
array slicing, transposing arrays, and checking data types.

```python
import numpy as np
```

## Array Basics

```python
fruits = ["apple", "banana", "cherry", "date"]
          
print(fruits)          
```

```python
arr_1 = np.array(fruits)

print(arr_1)
```

```python
print(arr_1.dtype)
```

The letter U means Unicode string. The number 6 means the maximum string length is 6 characters.

```python
arr_2 = np.array ([
    [10, 11, 12],
    [13, 14, 15]
])

print(arr_2)
```

```python
print(arr_2.T)
```

```python
print(arr_2)
```

arr_2 itself was not permanently changed | .T only displays a transposed version.

```python
arr_3 = np.array([
    [1, 3, 5],
    [2, 4, 6]
])

print(arr_3)
```

```python
arr_4 = np.array([
    arr_3[0, 2],
    arr_3[1, 1]
])

print(arr_4)
```

```python
arr_5 = np.ones((2, 2, 2))

print(arr_5)
```

arr_5 is a three dimensional array filled with ones. Each dimension contains 2 values.
