```python
# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings - part codes
```

```python
def get_supplier_code(code):
    return code.split(':')[0]
```

```python
def get_product_num(code):
    return code.split(':')[1].split('-')[0]
```

```python
def get_size(code):
    return code.split('-')[1]
```

```python
part1 = 'ACME:123-L'
part2 = 'SI:12345-M'
part3 = 'ACE:1-12'
```

```python
print(f"For {part1}, the supplier is {get_supplier_code(part1)}. Part number {get_product_num(part1)} is available in size {get_size(part1)}.")
```

```python
print(f"For {part2}, the supplier is {get_supplier_code(part2)}. Part number {get_product_num(part2)} is available in size {get_size(part2)}.")
```

```python
print(f"For {part3}, the supplier is {get_supplier_code(part3)}. Part number {get_product_num(part3)} is available in size {get_size(part3)}.")
```
