## Series

```python
aa = [1, 2, 3, 4]
ab = [5, 6, 7, 8]
ac = [9, 10, 11, 12]
ad = {
    'first': 13,
    'second': 14,
    'third': 15,
    'fourth': 16
}
```

```python
ser1 = pd.Series(aa)
ser2 = pd.Series(ab)
ser3 = pd.Series(ac)
ser4 = pd.Series(ad)

print(ser1)
print(ser2)
print(ser3)
print(ser4)
```

```python
ser5 = pd.Series(17, 18, 19, 20)
```

This gives a TypeError because pd.Series() does not take multiple seperate literal values like that. The error messafe is: Series.__init__() takes from 1 to 4 positional arguments but 5 were given.

## DataFrames from Series

```python
df1 = pd.DataFrame(ser1)
df1
```

```python
df1.insert(1, "ser2", ser2)
df1.insert(2, "ser3", ser3)
df1
```

```python
df1.insert(3, "ser4", ser4)
df1
```

ser4 came from a dictionary, so it has labels like first, second, third, and fourth instead of the same number index as the other Series. Mainly because the indexes do not match, pandas fills the column with NaN values.

```python
df1.loc[4] = [5, 9, 12, 0]
df1
```
