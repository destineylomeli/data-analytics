# Exercise 3.A Getting familiar with Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Creating Basic Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='o')

plt.title('Example Graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()
```

## Styling Plots

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, linestyle='--', color='red')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='s', markersize=10, markerfacecolor='green')

plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.plot(x, y, marker='^', color='blue')

plt.show()
```

# Exercise 3.B More plotting practice

## Advanced Plotting Techniques

```python
# Bar chart
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales)

plt.show()
```

```python
# Customized Bar Chart
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [23, 17, 35, 29]

plt.bar(products, sales, color='skyblue', width=0.5)

plt.title('Product Sales')
plt.xlabel('Products')
plt.ylabel('Sales')

plt.show()
```

```python
# Basic Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y)

plt.show()
```

```python
# Customized Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
sizes = np.random.rand(50) * 100
colors = np.random.rand(50)

plt.scatter(x, y, s=sizes, c=colors, cmap='viridis')

plt.title('Scatter Plot Example')
plt.xlabel('X values')
plt.ylabel('Y values')

plt.show()
```

## Subplots

```python
# Subplots With Line Chart And Histogram
fig, ax = plt.subplots(1, 2, figsize=(10, 4))

# First Subplot
ax[0].plot(x, y)
ax[0].set_title('Sine Wave')
ax[0].set_xlabel('X values')
ax[0].set_ylabel('Y values')

# Second Subplot
random_numbers = np.random.randn(1000)

ax[1].hist(random_numbers)
ax[1].set_title('Histogram')
ax[1].set_xlabel('Values')
ax[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
