```python
import numpy as np
```

## Sample Arrays

```python
r_1 = np.random.randint(1, 20, size=2)
print(r_1)
```

```python
r_2 = np.random.randint(1, 20, size=(1, 2))
print(r_2)
```

```python
r_3 = np.random.rand(2, 1)
print(r_3)
```

```python
r_4 = np.random.randint(20, 100, size=(4, 5))
print(r_4)
```

```python
r_5 = np.random.rand(4, 5)
print(r_5)
```

## Permutations

```python
print(r_4)
```

```python
np.random.shuffle(r_4)
print(r_4)
```

This rows of the array were shuffled into a different order.

```python
new_shuffle = np.random.permutation(r_4)
print(new_shuffle)

print(r_4)
```

Shuffle changes the original array directly. Permulation creates a shuffled copy while keeping the original array unchanged.

# Mathematical Operations

```python
print(r_1)
print(r_1.shape)

print(r_2)
print(r_2.shape)

result_1 = np.add(r_1, r_2)
print(result_1)
print(result_1.shape)
```

```python
print(r_1)
print(r_1.shape)

print(r_3)
print(r_3.shape)

result_2 = np.add(r_1, r_3)
print(result_2)
print(result_2.shape)
```

```python
print(r_2)
print(r_2.shape)

print(r_4)
print(r_4.shape)

result_3 = np.add(r_2, r_4)
print(result_3)
print(result_3.shape)
```

This will create an error because the array shapes are incompatible.

```python
np.add(r_1, r_4)
```

This will also create an error because the shapes do not match correctly.


```python
np.add(r_2, r_3)
```

```python
print(np.add(np.array([1]), r_4))
```

```python
print(np.add(np.array([[1]]), r_4))
```

```python
print(np.add(np.array([1,1,1,1,1]), r_4))
```

```python
np.add(np.array([[1],[1]]), r_4)
```

This produces an error because the shapes cannot broadcast together.

Broadcasting works by matching array shapes together. Values are added element by element when the dimensions are compatible. The resulting array shape is based on the larger compatible dimensions.

# Grocery Problem

```python
bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
total_bag1 = np.sum(bag1)
total_bag2 = np.sum(bag2)

print(f"The first bag contains {total_bag1} items, and the second bag contains {total_bag2} items.")
```

```python
totals = np.add(bag1, bag2)
print(f"There are {totals[0]} vegetables, {totals[1]} fruits, {totals[2]} boxes of snacks, and {totals[3]} beverages in the two grocery bags.")
```

# Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
print(ali)
```

```python
ben = np.array([0, 5, 5, 3, 4])
print(ben)
```

```python
carla = np.array([3, 6, 0, 3, 4])
print(carla)
```

```python
all_knocks = np.array([ali, ben, carla])
print(all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1)
group_total = np.sum(all_knocks)

print(person_totals)
print(group_total)
```

```python
daily_totals = np.sum(all_knocks, axis=0)

rolling_votes = np.cumsum(daily_totals) * 0.5

print(rolling_votes)
```
