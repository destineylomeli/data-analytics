```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("test_data.csv")
df
```

```python
df.describe()
```

## Filtering a DataFrame

```python
df[df['Age'] > 18]
```

```python
df[(df['Score'] > 50) & (df['Score'] < 100)]
```

## Modifying a DataFrame

```python
df_cleaned = df.drop(df[df['Score'] < 0].index)
df_cleaned
```

```python
df_cleaned = df_cleaned.drop(columns=['City'])
df_cleaned
```

```python
mean_age = df_cleaned['Age'].mean()
df_cleaned['Age'] = df_cleaned['Age'].fillna(mean_age)

df_cleaned
```

```python
mean_score = df_cleaned['Score'].mean()
df_cleaned['Score'] = df_cleaned['Score'].replace(0, mean_score)

df_cleaned
```

```python
df_cleaned = df_cleaned.dropna()
df_cleaned
```

```python
df_cleaned = df_cleaned.set_index('ID')
df_cleaned
```
