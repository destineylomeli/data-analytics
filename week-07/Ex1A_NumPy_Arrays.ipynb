# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13],
                [14, 15, 16]])
print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

The first loop multiplies each row by 2 as a group. The second loop goes thourgh each individual value and multiplies each number by 2 seperately.

```python
for x in arrA:
    for y in x:
        loop_result = y * 2
        print(loop_result)
```

The results contain the same values, but the first loop works with entire rows of the array while the second loop works with each individual number separetly. The first loop prints arrays, and the second loop prints one value at a time.

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)
print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)
print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))
print(stacked_array_1)
```

```python
# This will create an error because the arrays do not have matching shapes.
stacked_array_2 = np.stack((joined_array_1, joined_array_2))
print(stacked_array_2)
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))
print(another_joined_array)
```

```python
# This creates an error because the number of columns does not match for axis=1.
another_2 = np.concatenate((joined_array_1, arrA), axis = 1)
print(another_2)
```

```python
horizontal_stack = np.hstack((arrA, arrA))
print(horizontal_stack)
```

hstack joins arrays side by side horizontally.

```python
vertical_stack = np.vstack((arrA, arrA))
print(vertical_stack)
```

vstack joins arrays on top of each other vertically.

```python
depth_stack = np.dstack((arrA, arrA))
print(depth_stack)
```

dstack stacks arrays by depth, creating a third dimension.

# Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)
print(splitsky)
```

This splits the array into 2 smaller arrays. Each array still have 2 dimensions.

```python
spliterino = np.array_split(joined_array_1, 3)
print(spliterino)
```

This splits the array into 3 arrays. Each array is still 2 dimensional, but the rows are unevenl divided.

# Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')
print(result_1)
```

# Sorting and Filtering Arrays

```python
arrB = np.array([[7, 6],
                 [13, 5],
                 [7, 3]])
arrB_sorted = np.sort(arrB)
print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7
arrC = arrB[z]

print(arrC)
```

```python
z = arrB >= 7
arrD = arrB[z]

print(arrD)
```
